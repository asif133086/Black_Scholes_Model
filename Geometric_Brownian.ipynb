```python
import numpy as np
import matplotlib.pyplot as plt

S0 = 100     # initial stock price
mu = 0.05    # drift
sigma = 0.2  # volatility
T = 1.0      # time horizon (1 year)
N = 252      # number of steps (daily)
dt = T / N

t = np.linspace(0, T, N)
S = np.zeros(N)
S[0] = S0

for i in range(1, N):
    Z = np.random.normal()
    S[i] = S[i-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)

plt.plot(t, S)
plt.title("Geometric Brownian Motion Path")
plt.xlabel("Time (t)")
plt.ylabel("Stock Price (S)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

S0 = 100       # Initial stock price
mu = 0.05      # Drift
sigma = 0.2    # Volatility
T = 1.0        # Time horizon (1 year)
N = 252        # Number of time steps (daily)
M = 10         # Number of GBM paths
dt = T / N     # Time step size

t = np.linspace(0, T, N)

paths = np.zeros((M, N))
paths[:, 0] = S0

for m in range(M):
    for i in range(1, N):
        Z = np.random.normal()
        paths[m, i] = paths[m, i-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)

plt.figure(figsize=(10, 6))
for m in range(M):
    plt.plot(t, paths[m])
plt.title("Multiple Geometric Brownian Motion (GBM) Paths")
plt.xlabel("Time (Years)")
plt.ylabel("Stock Price")
plt.grid(True)
plt.savefig('gbm.png')
plt.show()
```
